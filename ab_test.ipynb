```python
import numpy as np
import matplotlib.pyplot as plt
from plots import *
set_seaborn()

t = np.linspace(0,5,30)
y = np.sin(t) - 100
y2 = .3*y+2+ 0.01 * np.random.normal(size=len(t))

plt.plot(t,y); plt.plot(t,y2);

# solve for scale and bias
y_hat = np.vstack((y2, np.ones(len(y2))))
ab = (y @ y_hat.T) @ np.linalg.inv((y_hat @ y_hat.T))
a, b = ab[0], ab[1]

print(a,b)
print(y2.shape)
plt.plot(t, a*y2+b, '*')

plt.show()
```
